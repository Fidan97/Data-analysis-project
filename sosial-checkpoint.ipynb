Introduction:


***This dataset contains social media posts from multiple platforms, capturing details such as post type, topic, posting time, and location. It also includes metrics on audience interaction, such as likes, comments, shares, saves, profile visits, follows, and link clicks.

The main objective of this analysis is to explore patterns in post performance, identify which topics and post types generate higher engagement, and determine the optimal posting times and locations. Insights derived from this dataset can help guide content strategy and improve overall audience interaction on social media platforms.***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sosial=pd.read_csv('social_media_post_performance_100.csv')
sosial.head()
```

```python
sosial['date'] = pd.to_datetime(sosial['date'], format='%Y-%m-%d')
```

```python
sosial.columns
```

```python
sosial.isnull().sum()
```

```python
sosial['saves'] = pd.to_numeric(sosial['saves'], errors='coerce')
sosial['saves'].fillna(sosial['saves'].mean(),inplace=True)
```

```python
sosial['save_rate'] = pd.to_numeric(sosial['save_rate'], errors='coerce')
sosial['save_rate'].fillna(sosial['save_rate'].mean(),inplace=True)
```

```python
sosial['share_rate'] = pd.to_numeric(sosial['share_rate'], errors='coerce')
sosial['share_rate'].fillna(sosial['share_rate'].mean(),inplace=True)
```

```python
sosial['shares'].fillna(sosial['shares'].mean(),inplace=True)
```

```python
sosial.isnull().sum()
```

```python
topic_counts = sosial.groupby('city')['topic'].count()
topic_counts.plot(kind='barh', color='blue')
plt.xlabel("Number of Topics")
plt.ylabel("City")
plt.title("Number of Topics per City")
plt.show()
```

```python
p_counts=sosial['post_type'].value_counts()
p_counts
```

```python
plt.pie(p_counts,labels=p_counts.index,autopct='%1.1f%%', colors=['skyblue','yellow','red','orange'])
plt.title("Distribution of Post Types")
plt.show()
```

```python
hour_counts = sosial['hour'].value_counts().sort_index()
hour_counts.plot(kind='bar', color='skyblue')
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Posts")
plt.title("Post Activity by Hour")
plt.show()
```

**Finding highest engagement**

```python
top_posts = sosial.sort_values(by='engagement_rate', ascending=False).head(10)
top_posts
```

```python
low_posts = sosial.sort_values(by='engagement_rate', ascending=True).head(10)
low_posts
```

```python
sosial.columns
```

```python
sosial['total_engagement'] = sosial[['likes','comments','shares','saves']].sum(axis=1)
sosial.head()
```

```python
top_posts2 = sosial.sort_values(by='total_engagement', ascending=False).head(10)
top_posts2
```

```python
low_posts2 = sosial.sort_values(by='total_engagement', ascending=True).head(10)
low_posts2
```

```python
sns.barplot(x='total_engagement', y='post_id', data=low_posts2, palette='coolwarm')
plt.xlabel("Total Engagement")
plt.ylabel("Post ID")
plt.title("Bottom 10 Posts by Total Engagement")
plt.show()
```

***The bottom 10 posts, based on total engagement, show significantly lower interaction compared to other posts, indicating areas where content or timing could be improved to increase audience engagement.***


Post 1: On 2025-07-01, a TikTok Reel about Python was posted in Baku at 19:00. The post achieved 16,096 impressions, 12,397 reach, 521 likes, 110 comments, and had a total engagement of 2,416. Engagement rates indicate strong interaction with the audience.
Python posts in the evening (19:00) seem to get higher engagement than SQL posts in the morning (9:00).
